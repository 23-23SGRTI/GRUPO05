```python
import timeit
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

def cifrar_archivo(ruta_archivo, clave):
    backend = default_backend()

    # Medir tiempo de lectura del archivo
    inicio_lectura = timeit.default_timer()

    with open(ruta_archivo, 'r') as archivo:
        mensaje = archivo.read()

    tiempo_lectura = (timeit.default_timer() - inicio_lectura) * 1000  # Convertir a milisegundos

    # Medir tiempo de cifrado
    inicio_cifrado = timeit.default_timer()

    mensaje_cifrado = ''

    # Aplicar algoritmo de cifrado de sustitución
    for caracter in mensaje:
        if caracter.isalpha():
            indice = (ord(caracter) - ord('a') + clave) % 26
            caracter_cifrado = chr(indice + ord('a'))
        else:
            caracter_cifrado = caracter

        mensaje_cifrado += caracter_cifrado

    tiempo_cifrado = (timeit.default_timer() - inicio_cifrado) * 1000  # Convertir a milisegundos

    ruta_archivo_cifrado = ruta_archivo + '.cifrado'
    with open(ruta_archivo_cifrado, 'w') as archivo_cifrado:
        archivo_cifrado.write(mensaje_cifrado)

    return ruta_archivo_cifrado, tiempo_lectura, tiempo_cifrado

def descifrar_archivo(ruta_archivo_cifrado, clave):
    backend = default_backend()

    # Medir tiempo de lectura del archivo cifrado
    inicio_lectura = timeit.default_timer()

    with open(ruta_archivo_cifrado, 'r') as archivo_cifrado:
        mensaje_cifrado = archivo_cifrado.read()

    tiempo_lectura = (timeit.default_timer() - inicio_lectura) * 1000  # Convertir a milisegundos

    # Medir tiempo de descifrado
    inicio_descifrado = timeit.default_timer()

    mensaje_descifrado = ''

    # Aplicar algoritmo de descifrado de sustitución
    for caracter in mensaje_cifrado:
        if caracter.isalpha():
            indice = (ord(caracter) - ord('a') - clave) % 26
            caracter_descifrado = chr(indice + ord('a'))
        else:
            caracter_descifrado = caracter

        mensaje_descifrado += caracter_descifrado

    tiempo_descifrado = (timeit.default_timer() - inicio_descifrado) * 1000  # Convertir a milisegundos

    ruta_archivo_descifrado = ruta_archivo_cifrado + '.descifrado'
    with open(ruta_archivo_descifrado, 'w') as archivo_descifrado:
        archivo_descifrado.write(mensaje_descifrado)

    return ruta_archivo_descifrado, tiempo_lectura, tiempo_descifrado

# Ejemplo de uso
clave_sustitucion = 3
ruta_archivo = r"C:/Users/luis1/OneDrive/Escritorio/Palabras/1000000.txt"


ruta_archivo_cifrado, tiempo_lectura_cifrado, tiempo_cifrado = cifrar_archivo(ruta_archivo, clave_sustitucion)
print("Archivo cifrado:", ruta_archivo_cifrado)
print("Tiempo de lectura del archivo cifrado:", tiempo_lectura_cifrado, "milisegundos")
print("Tiempo de cifrado:", tiempo_cifrado, "milisegundos")

ruta_archivo_descifrado, tiempo_lectura_descifrado, tiempo_descifrado = descifrar_archivo(ruta_archivo_cifrado, clave_sustitucion)
print("Archivo descifrado:", ruta_archivo_descifrado)
print("Tiempo de lectura del archivo descifrado:", tiempo_lectura_descifrado, "milisegundos")
print("Tiempo de descifrado:", tiempo_descifrado, "milisegundos")
```
