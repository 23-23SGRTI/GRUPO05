```python
import time
from cryptography.hazmat.primitives import serialization, hashes
from cryptography.hazmat.primitives.asymmetric import rsa, padding
```

```python
# Leer el archivo con el texto a cifrar
file_name = r"C:/Users/luis1/OneDrive/Escritorio/Palabras/10.txt"

with open(file_name, 'rb') as file:
    message  = file.read()
```

```python
# Generar un par de claves RSA
start_time = time.time()

private_key = rsa.generate_private_key(
    public_exponent=65537,
    key_size=2048
)
public_key = private_key.public_key()

end_time = time.time()
key_generation_time = (end_time - start_time) * 1000  # Convertir a milisegundos
```

```python
# Serializar y guardar las claves en archivos
start_time = time.time()

pem_private = private_key.private_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PrivateFormat.PKCS8,
    encryption_algorithm=serialization.NoEncryption()
)
pem_public = public_key.public_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PublicFormat.SubjectPublicKeyInfo
)

with open('private_key.pem', 'wb') as f:
    f.write(pem_private)

with open('public_key1.pem', 'wb') as f:
    f.write(pem_public)

end_time = time.time()
key_saving_time = (end_time - start_time) * 1000  # Convertir a milisegundos

```

```python
# Cifrado
start_time = time.time()

ciphertext = public_key.encrypt(
    message,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA512()),
        algorithm=hashes.SHA512(),
        label=None
    )
)

end_time = time.time()
encryption_time = (end_time - start_time) * 1000  # Convertir a milisegundos
```

```python
start_time = time.time()

plaintext = private_key.decrypt(
    ciphertext,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA512()),
        algorithm=hashes.SHA512(),
        label=None
    )
 )

# Comprobamos que sean el mismo dato
plaintext == message
True

end_time = time.time()
decryption_time = (end_time - start_time) * 1000  # Convertir a milisegundos


```

```python

# Mostrar los resultados y tiempos transcurridos
print("Texto a cifrar:")
print(message.decode('utf-8'))
print()

print("Claves de cifrado:")
print(pem_private.decode('utf-8'))
print(pem_public.decode('utf-8'))
print()

print("Texto cifrado:")
print(ciphertext)
print()

print("Texto descifrado:")
print(plaintext.decode('utf-8'))
print()

print("Tiempos transcurridos:")
print("Generación de claves:", key_generation_time, "milisegundos")
print("Guardado de claves:", key_saving_time, "milisegundos")
print("Cifrado:", encryption_time, "milisegundos")
print("Descifrado:", decryption_time, "milisegundos")
```
